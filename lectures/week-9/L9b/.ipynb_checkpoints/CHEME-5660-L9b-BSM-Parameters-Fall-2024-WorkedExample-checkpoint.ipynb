```julia
Î”t = (1.0/365.0); # we are going to use calendar years - this is DIFF than Hull (uses trade years)
Sâ‚’ = 60.0; # inital share price
K = 60.0; # strike price
T = 365.0*Î”t; # duration of the contract (days to expiration)
rÌ„ = 0.05; # risk free rate
ÏƒÌ„ = 0.10; # implied volatility (market expectation of future volatility)
ð’Ÿ(r,T) = exp(r*T); # discount factor
```

## Task 1: How does time to expiration influence contract premium?
In this task, we'll explore how the time to expiration $T$ influences the premium of European option contracts. In particular, we'll look at two contracts: 
* __Contract 1__: A European `call` that expires in `T = 365` days with a strike price of `K = 60.0 USD`. The current share price is `S(0) = 60.0 USD`. Assume a risk-free rate of return of `5%` and an implied volatility of `10%`. The current premium is $\mathcal{P}_{c}(K,S(0))$ = `4.08 USD/share`.
* __Contract 2__: A European `put` that expires in `T = 365` days with a strike price `K = 60.0 USD`. The current share price is `S(0) = 60.0 USD`. Assume a risk-free rate of return of `5%` and an implied volatility of `10%`. The current answer is $\mathcal{P}_{p}(K,S(0))\simeq$ `1.16 USD/share`.

First, let's build models for the contracts that hold the default contract parameters (we'll then update these models for different values of the parameters in the calculations below):

```julia
call_contract_model, put_contract_model = let

    # specified shared parameters -
    DTE = T;
    sense = 1;
    copynumber = 1;
    IV = 0.1;

    # build the call model -
    call_contract_model = build(MyEuropeanCallContractModel, (
        K = K, IV = IV, DTE = DTE, sense = sense, copy = copynumber));

    # build the put model -
    put_contract_model = build(MyEuropeanPutContractModel, (
        K = K, IV = IV, DTE = DTE, sense = sense, copy = copynumber));

    # return -
    call_contract_model, put_contract_model
end;
```

Next, let create an instance of [the `MyBlackScholesContractPricingModel` type](https://varnerlab.github.io/VLQuantitativeFinancePackage.jl/dev/derivatives/#VLQuantitativeFinancePackage.MyBlackScholesContractPricingModel), which holds the risk-free rate $\bar{r}$ and the initial share price $S_{\circ}$. We store this instance in the `bsm_model::MyBlackScholesContractPricingModel` variable:

```julia
bsm_model = build(MyBlackScholesContractPricingModel, (
        Sâ‚’ = Sâ‚’, r = rÌ„
));
```

Now that we have contract models and the best model, we can simulate what happens to the premium of a call or put option if the time to expiration $T$ changes (all other parameters held constant).

```julia
DTE_simulation_df = let

    df = DataFrame()
    for Î²áµ¢ âˆˆ Î² # DTE
        Táµ¢ = Î²áµ¢*T;

        call = deepcopy(call_contract_model);
        call.DTE = Táµ¢; # set the new DTE on the call

        put = deepcopy(put_contract_model);
        put.DTE = Táµ¢; # set the new DTE on the put
        
        for Kâ±¼ âˆˆ KA # strike
            
            call.K = Kâ±¼ # set the new strike on the call
            put.K = Kâ±¼ # set the next strike on the put

            # capture -
            row_df_call = (
                DTE = Táµ¢,
                K = Kâ±¼,
                scaled_price = log(Sâ‚’/Kâ±¼),
                contract = :call,
                premium = premium(call, bsm_model)
            );
            push!(df, row_df_call);

            row_df_put = (
                DTE = Táµ¢,
                K = Kâ±¼,
                scaled_price = log(Sâ‚’/Kâ±¼),
                contract = :put,
                premium = premium(put, bsm_model)
            );
            push!(df, row_df_put);
        end
    end

    df
end;
```

```julia
DTE_simulation_df
```

### Visualize call option pricing
`Unhide` the block below to see how we plot the European call premium as a function of the scaled share price and the time to expiration. For a long call, a buyer who exercises thier contract buys at the strike $K$ and sells at the market $S$.

#### Pricing regimes
Let's look at three pricing regimes.
* __Regime I__: $K\rightarrow{\infty}$. For this case, the initial share price $S_{\circ}\ll{K}$.
The premium of the European call goes to zero as $K\rightarrow{\infty}$ for all time to expiration values.
* __Regime II__: $K\approx{S_{\circ}}$. For this case, the premium of the European call contract is directly proportional to the duration of the contract, i.e., long-duration contracts are more expensive, and contracts have non-zero premiums even in the case $S_{\circ}<{K}$.
* __Regime III__: $K\rightarrow{0}$. For this case, the initial share price $S_{\circ}\gg{K}$.
The premium of the European call goes to $\infty$ as $K\rightarrow{0}$ for all times to expiration values. However, longer dates contracts are more expensive.

```julia
let
    q = plot();
    for Î²áµ¢ âˆˆ Î² # DTE
        Táµ¢ = Î²áµ¢*T;
        df = filter([:DTE, :contract] => (x,y) -> x == Táµ¢ && y == :call,  DTE_simulation_df)
        plot!(df[:,:K], df[:, :premium], label="", c=:gray80)
    end

    # payoff -
    payoff = Array{Float64,2}(undef, length(KA),2);
    for i âˆˆ eachindex(KA)
        payoff[i,1] = KA[i];
        payoff[i,2] = max(Sâ‚’-KA[i],0); # payoff today of the contract
    end

    # plot the payoff -
    plot!(payoff[:,1], payoff[:,2], c=:tan1, label="Payoff (today)", lw=2)
    
    # plot -
    T1 = Î²[1]*T;
    df = filter([:DTE, :contract] => (x,y) -> x == T1 && y == :call,  DTE_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:red, lw=3, label="T = $(T1) years")

    df = filter([:DTE, :contract] => (x,y) -> x == T && y == :call,  DTE_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:navy, lw=3, label="T = $(T) years")

    T2 = Î²[end]*T;
    df = filter([:DTE, :contract] => (x,y) -> x == T2 && y == :call,  DTE_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:lightskyblue, lw=3, label="T = $(T2) years")

    vertical_line_y = range(0,step=0.1,stop=24) |> collect;
    vertical_line_x = Sâ‚’*ones(length(vertical_line_y));
    plot!(vertical_line_x,vertical_line_y, lw=2, c=:gray75, ls=:dash, label="Sâ‚’ = 60.0 (today)")
    xlabel!("Strike price K (USD/share)", fontsize=18)
    ylabel!("Contract premium call (USD/share)", fontsize=18)
    current()
end
```

## Task 2: How does volatility influence contract premium?
In this task, let's look at how the volatility of the underlying asset influences the premium of European options contracts. We'll look at the same two contracts from before:
* __Contract 1__: A European `call` that expires in `T = 365` days with a strike price of `K = 60.0 USD`. The current share price is `S(0) = 60.0 USD`. Assume a risk-free rate of return of `5%` and an implied volatility of `10%`. The current premium is $\mathcal{P}_{c}(K,S(0))$ = `4.08 USD/share`.
* __Contract 2__: A European `put` that expires in `T = 365` days with a strike price `K = 60.0 USD`. The current share price is `S(0) = 60.0 USD`. Assume a risk-free rate of return of `5%` and an implied volatility of `10%`. The current answer is $\mathcal{P}_{p}(K,S(0))\simeq$ `1.16 USD/share`.

```julia
volatility_simulation_df = let

    df = DataFrame()
    for Î²áµ¢ âˆˆ Î² # Ïƒ -
        Ïƒáµ¢ = Î²áµ¢*ÏƒÌ„;
        
        for Kâ±¼ âˆˆ KA # strike
            call = deepcopy(call_contract_model);
            call.IV = Ïƒáµ¢
            call.K = Kâ±¼

            put = deepcopy(put_contract_model);
            put.IV = Ïƒáµ¢;
            put.K = Kâ±¼

            # capture -
            row_df_call = (
                IV = Ïƒáµ¢,
                K = Kâ±¼,
                scaled_price = log(Sâ‚’/Kâ±¼),
                contract = :call,
                premium = premium(call, bsm_model)
            );
            push!(df, row_df_call);

            row_df_put = (
                IV = Ïƒáµ¢,
                K = Kâ±¼,
                scaled_price = log(Sâ‚’/Kâ±¼),
                contract = :put,
                premium = premium(put, bsm_model)
            );
            push!(df, row_df_put);
        end
    end

    df
end;
```

### Visualize

```julia
let
    q = plot();
    for Î²áµ¢ âˆˆ Î² # DTE
        Ïƒáµ¢ = Î²áµ¢*ÏƒÌ„;
        df = filter([:IV, :contract] => (x,y) -> x == Ïƒáµ¢ && y == :call,  volatility_simulation_df)
        plot!(df[:,:K], df[:, :premium], label="", c=:gray80)
    end

    # payoff -
    payoff = Array{Float64,2}(undef, length(KA),2);
    for i âˆˆ eachindex(KA)
        payoff[i,1] = KA[i];
        payoff[i,2] = max(Sâ‚’-KA[i],0);
    end

    # plot the payoff -
    plot!(payoff[:,1], payoff[:,2], c=:tan1, label="Payoff", lw=2)

    # plot -
    Ïƒáµ¢ = Î²[1]*ÏƒÌ„;
    df = filter([:IV, :contract] => (x,y) -> x == Ïƒáµ¢ && y == :call,  volatility_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:red, lw=3, label="IV = $(Ïƒáµ¢ |> x-> round(x, digits=3))")

    df = filter([:IV, :contract] => (x,y) -> x == ÏƒÌ„ && y == :call,  volatility_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:navy, lw=3, label="IV = $(ÏƒÌ„ |> x-> round(x, digits=3))")

    Ïƒáµ¢ = Î²[end]*ÏƒÌ„;
    df = filter([:IV, :contract] => (x,y) -> x == Ïƒáµ¢ && y == :call,  volatility_simulation_df)
    plot!(df[:,:K], df[:, :premium], c=:lightskyblue, lw=3, label="IV = $(Ïƒáµ¢ |> x-> round(x, digits=3))")

    vertical_line_y = range(0,step=0.1,stop=24) |> collect;
    vertical_line_x = Sâ‚’*ones(length(vertical_line_y));
    plot!(vertical_line_x,vertical_line_y, lw=2, c=:gray75, ls=:dash, label="Sâ‚’ = 60.0 (today)")
    xlabel!("Strike price (USD/share)", fontsize=18)
    ylabel!("Contract premium call (USD/share)", fontsize=18)
    
    current()
end
```
